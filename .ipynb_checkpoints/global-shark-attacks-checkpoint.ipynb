```python
import chardet
import pandas as pd
import re
```

```python
with open('./input/global_shark_attacks.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large


df = pd.read_csv('./input/global_shark_attacks.csv', encoding=result['encoding'])
```

```python
df.head(10)
```

```python
# I have realized that some columns are repeated or do not have any value for my analysis. I am going to remove them.

df = df.drop(columns=['Unnamed: 22','Unnamed: 23','original order','Case Number.2','Case Number.1','href','href formula','pdf','Investigator or Source',])
```

```python
df.head(10)
```

```python
short_df = df.reindex(columns=["Year",'Type','Country','Area','Activity','Sex','Age','Fatal (Y/N)','Time','Species'])
```

```python
#reunion = short_df[['Year','Area']].groupby[('Area')].agg(['count'])
count_areas = short_df.groupby(['Area']).size()
count_country = short_df.groupby(['Country']).size()
```

```python
display(count_areas.idxmax())
display(count_areas.max())
display(count_areas.idxmin())
display(count_areas.min())
```

```python
display(count_country.idxmax())
display(count_country.max())
display(count_country.idxmin())
display(count_country.min())
```

```python
areas_df = pd.DataFrame({'area':count_areas.index, 'attacks':count_areas.values})
country_df = pd.DataFrame({'country':count_country.index, 'attacks':count_country.values})
```
